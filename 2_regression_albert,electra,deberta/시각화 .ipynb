```python
import matplotlib.pyplot as plt

epochs = [1, 2, 3, 4, 5, 6]
training_loss = [0.3958, 0.328500, 0.274, 0.1964, 0.127, 0.0762]
validation_loss = [0.345885, 0.356576, 0.369285, 0.341242, 0.359982, 0.385895]
qwk = [0.763585, 0.758873, 0.757875, 0.782262, 0.778958, 0.775631]
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Loss', color=color)
ax1.plot(epochs, training_loss, color=color, label='Training Loss')
ax1.plot(epochs, validation_loss, color=color, linestyle='--', label='Validation Loss')
ax1.tick_params(axis='y', labelcolor=color)
ax1.grid(True)
ax1.legend(loc='upper left')

ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('QWK', color=color)
ax2.bar(epochs, qwk, color=color, alpha=0.5, label='QWK', width=0.4)
ax2.tick_params(axis='y', labelcolor=color)

# Adjust y-axis limits and ticks
ax2.set_ylim(0.5, 1)
ax2.set_yticks([0.5, 0.55, 0.6, 0.65, 0.7, 0.75, 0.8, 0.85, 0.9, 0.95, 1.0])

ax2.legend(loc='upper right')

fig.tight_layout()
plt.title('Training and Validation Losses with QWK')
plt.show()
```
